```python
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

matrix_a_input = widgets.Textarea(
    value='1,2\n3,4',
    placeholder='Enter rows separated by new lines, values separated by commas',
    description='Matrix A:',
    layout=widgets.Layout(width='300px', height='100px')
)

matrix_b_input = widgets.Textarea(
    value='5,6\n7,8',
    placeholder='Enter rows separated by new lines, values separated by commas',
    description='Matrix B:',
    layout=widgets.Layout(width='300px', height='100px')
)

operation_dropdown = widgets.Dropdown(
    options=['Addition', 'Subtraction', 'Multiplication', 'Transpose A', 'Transpose B', 'Determinant A', 'Determinant B'],
    description='Operation:'
)

calculate_button = widgets.Button(description='Calculate', button_style='success')
output_area = widgets.Output()
```

```python
def parse_matrix(text):
    rows = text.strip().split('\n')
    matrix = []
    for row in rows:
        values = [float(x.strip()) for x in row.split(',')]
        matrix.append(values)
    return np.array(matrix)


def on_calculate_clicked(b):
    with output_area:
        clear_output()
        try:
            operation = operation_dropdown.value

            if operation in ['Addition', 'Subtraction', 'Multiplication']:
                a = parse_matrix(matrix_a_input.value)
                b_matrix = parse_matrix(matrix_b_input.value)

                if operation == 'Addition':
                    if a.shape != b_matrix.shape:
                        print('Error: Matrices must have the same dimensions for addition.')
                        return
                    result = a + b_matrix
                elif operation == 'Subtraction':
                    if a.shape != b_matrix.shape:
                        print('Error: Matrices must have the same dimensions for subtraction.')
                        return
                    result = a - b_matrix
                elif operation == 'Multiplication':
                    if a.shape[1] != b_matrix.shape[0]:
                        print('Error: Number of columns in A must equal number of rows in B.')
                        return
                    result = a @ b_matrix

                print(f'Matrix A:\n{a}\n')
                print(f'Matrix B:\n{b_matrix}\n')
                print(f'Result ({operation}):\n{result}')

            elif operation == 'Transpose A':
                a = parse_matrix(matrix_a_input.value)
                print(f'Matrix A:\n{a}\n')
                print(f'Transpose of A:\n{a.T}')

            elif operation == 'Transpose B':
                b_matrix = parse_matrix(matrix_b_input.value)
                print(f'Matrix B:\n{b_matrix}\n')
                print(f'Transpose of B:\n{b_matrix.T}')

            elif operation == 'Determinant A':
                a = parse_matrix(matrix_a_input.value)
                if a.shape[0] != a.shape[1]:
                    print('Error: Determinant requires a square matrix.')
                    return
                print(f'Matrix A:\n{a}\n')
                print(f'Determinant of A: {np.linalg.det(a):.4f}')

            elif operation == 'Determinant B':
                b_matrix = parse_matrix(matrix_b_input.value)
                if b_matrix.shape[0] != b_matrix.shape[1]:
                    print('Error: Determinant requires a square matrix.')
                    return
                print(f'Matrix B:\n{b_matrix}\n')
                print(f'Determinant of B: {np.linalg.det(b_matrix):.4f}')

        except Exception as e:
            print(f'Error: {e}')
            print('Check your input format. Example: 1,2\n3,4')
```

```python
calculate_button.on_click(on_calculate_clicked)

print('Matrix Operations Tool')
print('Enter each row on a new line, separate values in a row with commas.')
print('Example for a 2x2 matrix:\n1,2\n3,4\n')

display(matrix_a_input, matrix_b_input, operation_dropdown, calculate_button, output_area)
```
